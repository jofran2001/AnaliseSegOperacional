```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid", palette="viridis")
print("Bibliotecas importadas com sucesso!")
```

```python
# Lista de valores que devem ser considerados como Nulos (NaN)
valores_nulos = ['***', 'NULL', '****', '*****']

# Carregando os principais arquivos CSV com o tratamento adequado
df_ocorrencias = pd.read_csv('ocorrencia.csv', sep=';', encoding='latin-1', na_values=valores_nulos)
df_aeronaves = pd.read_csv('aeronave.csv', sep=';', encoding='latin-1', na_values=valores_nulos)
df_ocorrencia_tipo = pd.read_csv('ocorrencia_tipo.csv', sep=';', encoding='latin-1', na_values=valores_nulos)

print("Dados carregados. Verificando as primeiras linhas de cada tabela:")
print("\nOcorrências:")
display(df_ocorrencias.head())

print("\nAeronaves:")
display(df_aeronaves.head())
```

```python
# --- TRATAMENTO DE DATAS ---
# Convertendo a coluna de data para o formato datetime, que permite análises temporais
# O errors='coerce' transforma datas inválidas em NaT (Not a Time), evitando que o código quebre
df_ocorrencias['ocorrencia_dia'] = pd.to_datetime(df_ocorrencias['ocorrencia_dia'], format='%d/%m/%Y', errors='coerce')

# Removendo linhas onde a data não pôde ser convertida (se houver)
df_ocorrencias.dropna(subset=['ocorrencia_dia'], inplace=True)

# Criando uma coluna 'ano' para facilitar o agrupamento
df_ocorrencias['ano'] = df_ocorrencias['ocorrencia_dia'].dt.year

print("Coluna 'ocorrencia_dia' convertida para data e coluna 'ano' criada.")


# --- MERGE (UNIÃO) DOS DADOS ---
# CORREÇÃO: Especificando as chaves corretas para a união dos dataframes
# Usamos 'left_on' para a coluna em df_ocorrencias e 'right_on' para a coluna em df_aeronaves
df_analise = pd.merge(
    df_ocorrencias,
    df_aeronaves,
    left_on='codigo_ocorrencia', # Chave do dataframe da esquerda
    right_on='codigo_ocorrencia2', # Chave do dataframe da direita
    how='left' # Mantém todas as ocorrências
)

print("\nDataFrames de ocorrências e aeronaves unidos com sucesso!")
display(df_analise.head())
```

```python
# Contando as ocorrências únicas por ano para evitar duplicatas do merge
ocorrencias_por_ano = df_analise.groupby('ano')['codigo_ocorrencia'].nunique()

# Plotando o gráfico
plt.figure(figsize=(14, 7))
sns.lineplot(x=ocorrencias_por_ano.index, y=ocorrencias_por_ano.values, marker='o', color='royalblue')
plt.title('Evolução do Número de Ocorrências Aeronáuticas por Ano no Brasil', fontsize=16)
plt.xlabel('Ano', fontsize=12)
plt.ylabel('Número de Ocorrências', fontsize=12)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.countplot(y=df_analise['ocorrencia_classificacao'],
              order=df_analise['ocorrencia_classificacao'].value_counts().index,
              palette='viridis')
plt.title('Distribuição por Classificação da Ocorrência', fontsize=16)
plt.xlabel('Contagem', fontsize=12)
plt.ylabel('Classificação', fontsize=12)
plt.show()
```

```python
# Vamos limpar um pouco o nome dos fabricantes para agrupar melhor (ex: 'EMBRAER S A' vira 'EMBRAER')
# Usaremos .str.upper() para padronizar para maiúsculas e .str.strip() para remover espaços
df_analise['aeronave_fabricante_limpo'] = df_analise['aeronave_fabricante'].str.upper().str.strip()

# Agrupando nomes comuns (exemplo)
df_analise['aeronave_fabricante_limpo'].replace('EMBRAER S A', 'EMBRAER', inplace=True)
df_analise['aeronave_fabricante_limpo'].replace('CESSNA AIRCRAFT', 'CESSNA', inplace=True)
df_analise['aeronave_fabricante_limpo'].replace('PIPER AIRCRAFT', 'PIPER', inplace=True)


# Contando os 20 principais fabricantes
top_20_fabricantes = df_analise['aeronave_fabricante_limpo'].value_counts().nlargest(20)

plt.figure(figsize=(12, 10))
sns.barplot(y=top_20_fabricantes.index, x=top_20_fabricantes.values, palette='plasma')
plt.title('Top 20 Fabricantes por Número de Ocorrências Registradas', fontsize=16)
plt.xlabel('Número de Ocorrências', fontsize=12)
plt.ylabel('Fabricante', fontsize=12)
plt.show()
```
